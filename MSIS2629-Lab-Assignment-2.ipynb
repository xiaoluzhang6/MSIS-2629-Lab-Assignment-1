

# Objectives:

Develop a visualizations that strongly support/refute the following claim:
    “Security breaches are  becoming less  severe”



```python
import pandas

# load data
df = pandas.read_csv('test_data.csv', delimiter=',')

# replace header with first row of data

new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```

```python
# drop unwanted columns
df.drop(df.columns[[ 1, 2,7,8,10,11,12,13,14,15,16,17]], axis=1,inplace=True)

# drop unwanted row
df.drop(df.index[0], inplace=True)

# change type of YEAR column to numeric and to actual value
df['YEAR'] = df['YEAR'].astype(int)
y = 2004
df['YEAR']+= y

df.reset_index()


```

```python
# check for empty/null fields

df.isnull()
```

```python
# check to see if there are invalid entries; row 135 has invalid 'DATA SENSITIVITY'entry; this corrupted data entry 
# will be deleted from the dataset

value_list = ['1','20','300','4000','50000']
df[~df['DATA SENSITIVITY'].isin(value_list)]
```

```python
# drop unwanted row. Since top row has been dropped in the beginning, index number of this row has become '133' instead
# of '134'

df.drop(df.index[133], inplace=True)

df.head(140)
```

```python
# filter out data; only want data entries associated with 'hacked'method of breach
hackedBreaches = df[(df['METHOD OF LEAK'] == 'hacked')]
```

```python
# export data in a csv file

hackedBreaches.to_csv('lab2_output.csv')
```

```python
# modifying table to filter data with high data sensitivity

hackedBreachesV2 = hackedBreaches.loc[df['DATA SENSITIVITY'] > '300']
```

```python
# export data in a csv file

hackedBreachesV2.to_csv('lab2_output_2.csv')
```
